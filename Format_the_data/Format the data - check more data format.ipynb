```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Format_the_data_check_data_types').getOrCreate()

df = spark.read.csv('../Bike-Sharing-Dataset-hour_new.csv',header=True,inferSchema=True)
```

```python
import pandas as pd
sum_df = df.select('reinstant', 'dteday', 'mnth', 'hr', 'temp', 'atemp', 'hum', 'windspeed', 'cnt').describe()
pd.DataFrame(sum_df.take(sum_df.count()), columns=sum_df.columns)
```

```python
#temp times 10 to check if the calculation is correct
sum_df = sum_df.withColumn('temp * 10',sum_df['temp']*10)
pd.DataFrame(sum_df.take(sum_df.count()), columns=sum_df.columns)
```

```python
#check random yr, mnth and cnt to retrieve expected results
fil_df = df.filter("yr = 0 AND mnth = 1 AND cnt < 2").select('reinstant','dteday','yr','mnth','hr','cnt')
pd.DataFrame(fil_df.take(5), columns=fil_df.columns)
```

```python
#check random date and hr to see if I can get the exact result
date_df = df.filter("dteday = '2011-02-04' AND hr = '8'").select('reinstant','dteday','yr','mnth','hr')
pd.DataFrame(date_df.take(date_df.count()), columns=date_df.columns)
```
