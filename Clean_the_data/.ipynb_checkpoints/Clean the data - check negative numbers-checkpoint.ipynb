```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Clean_the_data_check_negative_numbers').getOrCreate()
df = spark.read.csv('../Bike-Sharing-Dataset-hour.csv', header=True, inferSchema=True)
```

```python
df.createOrReplaceTempView("dataset")

#change the variable 'temp' accordingly to check other variables
sqlDF = spark.sql("SELECT count(temp) FROM dataset WHERE temp < 0 ")
sqlDF.show()
```
