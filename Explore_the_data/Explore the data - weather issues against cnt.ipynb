```python
import pandas
data = pandas.read_csv("../Bike-Sharing-Dataset-hour.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

hr = data["hr"][0:23]
hum = data["hum"][0:23]
temp = data["temp"][0:23]
windspeed = data["windspeed"][0:23]
cnt = data["cnt"][0:23]

plt.subplot(4, 1, 1)
plt.plot(hr, hum)
plt.title('Weather Issues and cnt on the First Day')
plt.ylabel('hr vs hum')

plt.subplot(4, 1, 2)
plt.plot(hr, temp)
plt.xlabel('')
plt.ylabel('hr vs temp')

plt.subplot(4, 1, 3)
plt.plot(hr, windspeed)
plt.xlabel('')
plt.ylabel('hr vs windspeed')

plt.subplot(4, 1, 4)
plt.plot(hr, cnt)
plt.xlabel('hr')
plt.ylabel('hr vs cnt')

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10)
```
