```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Explore_the_data_weathersit_vs_cnt').getOrCreate()
df = spark.read.csv('../Bike-Sharing-Dataset-hour.csv', header=True, inferSchema=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pyspark.sql.functions import mean

groupset = df.select('cnt','weathersit').groupBy('weathersit').mean()
groupset = groupset.toPandas()

plt.bar(groupset['weathersit'], groupset['avg(cnt)'])
plt.xlabel('weathersit')
plt.ylabel('average cnt')
plt.title('weathersit vs cnt')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15,10)
```
