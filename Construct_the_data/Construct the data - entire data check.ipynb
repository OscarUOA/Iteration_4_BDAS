```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Construct_the_data_entire_data_check').getOrCreate()
df = spark.read.csv('../Bike-Sharing-Dataset-hour.csv', header=True, inferSchema=True)
    
df.createOrReplaceTempView("dataset")
#query to count the data for the specific year and month
df = spark.sql("SELECT COUNT(dteday) FROM dataset ")
df.show()
```

```python
#query to retrieve the data for the specific year and month
df = spark.sql("SELECT * FROM dataset ORDER BY instant ")

import pandas as pd
pd.DataFrame(df.take(df.count()), columns=df.columns)
```
