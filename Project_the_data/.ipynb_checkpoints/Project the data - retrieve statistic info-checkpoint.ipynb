```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Project_the_data_statistic_info').getOrCreate()
```

```python
df = spark.read.csv('../Bike-Sharing-Dataset-hour_new.csv',header=True,inferSchema=True)
```

```python
stat_df = df.select('cnt', 'casual', 'registered', 'holiday', 'workingday', 
          'temp', 'atemp', 'hum', 'windspeed', 'weathersit').describe()

import pandas as pd
pd.DataFrame(stat_df.take(stat_df.count()), columns = stat_df.columns).transpose()
```
